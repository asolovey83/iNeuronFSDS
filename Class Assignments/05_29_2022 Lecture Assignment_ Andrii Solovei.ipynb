<b>Question 10:</b>  Write a code to trigger alarm for you at scheduled time

<b>Solution 10:</b>

```python
import sched, time, datetime

def print_time():
    print("The time is now: {}".format(datetime.datetime.now()))

# Run 10 seconds from now
when = time.time() + 10

# Create the scheduler
s = sched.scheduler(time.time)
s.enterabs(when, 1, print_time)

# Run the scheduler
print_time()
print("Executing s.run()")
s.run()
print("s.run() exited")
```

even in case of 2 level nesting it should work. 

<b>Solution 16:</b>

```python
def dict_to_list(mydict, mylist):
    
    for item in mydict:
        if type(mydict[item]) == dict:
            dict_to_list(mydict[item], mylist)
        else:
            mylist.append(mydict[item])
    return mylist
```

<b>Question 20:</b> Write a function by which you will be able to append two PDF files

<b>Solution 20:</b>

```python
!pip install PyPDF2
```

```python
from PyPDF2 import PdfFileMerger
```

```python
def merge_pdfs(pdf1, pdf2):
    
    merger = PdfFileMerger()
    merger.append(pdf1)
    merger.append(pdf2)
    merger.write("result.pdf")
    merger.close()
```

```python
merge_pdfs("Sample_pdf1.pdf", "Sample_pdf2.pdf")
```

<b>Question 21:</b> Write a function which can help you to filter only word file from a directory

<b>Solution 21:</b>

```python
def get_doc_files(dir_path):
    files = []
    docfiles = []
    
    for path in os.listdir(dir_path):
    
        if os.path.isfile(os.path.join(dir_path, path)):
            files.append(path)
        
    for file in files:
        if file.endswith(".doc") or file.endswith(".docx"):
            docfiles.append(file)
            
    print(docfiles)
```

```python
get_doc_files(os.getcwd())
```

<b>Question 22:</b> Write a function which can read video file and play for you

<b>Solution 22:</b>

```python
def play_video(filename):
    
    
    cap = cv2.VideoCapture(filename)
    ret, frame = cap.read()
    while(1):
        ret, frame = cap.read()
        cv2.imshow('frame',frame)
        if cv2.waitKey(1) & 0xFF == ord('q') or ret==False :
            cap.release()
            cv2.destroyAllWindows()
            break
        #cv2.imshow('frame',frame)    
```
